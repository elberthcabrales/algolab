```python
# https://leetcode.com/problems/bus-routes/

from collections import defaultdict, deque
from typing import List

class Solution:
    def numBusesToDestination(self, routes: List[List[int]], source: int, target: int) -> int:
        """
            map buses_stops
        """
        if source == target:
            return 0

        stops_buses = defaultdict(list)
        for bus, stops in enumerate(routes):
            for stop in stops:
                stops_buses[stop].append(bus)
        
        queue = deque([(source, 0)])
        buses_visited = set()
        stops_visited = set()

        while queue:
            current_stop, buses_taken = queue.popleft()

            if current_stop == target:
                return buses_taken
            
            for bus in stops_buses[current_stop]:
                if bus in buses_visited:
                    continue
                buses_visited.add(bus)

                for stop in routes[bus]:
                    if stop not in stops_visited:
                        queue.append((stop, buses_taken + 1))
                    stops_visited.add(stop)

        return -1
```

```python
# Refactored implementation of the graph_dfs_iterative function

def graph_dfs_iterative(graph, start_node):
    stack = [start_node]
    seen = set([start_node])
    count = 0  # Counter for nodes visited

    while stack:
        node = stack.pop()
        count += 1  # Process the current node

        for neighbor in graph[node]:
            if neighbor not in seen:
                seen.add(neighbor)
                stack.append(neighbor)

    return count

# Test case for the graph_dfs_iterative function
graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0, 3],
    3: [1, 2]
}

output_graph_dfs_iterative = graph_dfs_iterative(graph, 0)
output_graph_dfs_iterative
```

```python
# https://leetcode.com/problems/maximum-number-of-fish-in-a-grid

class Solution:
    def findMaxFish(self, grid: List[List[int]]) -> int:
        rows = len(grid)
        cols = len(grid[0])
        directions = [(0,1), (1,0), (-1,0), (0,-1)]
        isValid = lambda x, y: 0 <= x < rows and 0 <= y < cols and grid[x][y] > 0
        visited = set()
        queue = deque()

        fished = 0

        for r in range(rows):
            for c in range(cols):
                if grid[r][c] > 0:
                    queue = deque([(r, c)])
                    visited.add((r, c))
                    current = 0
                    while queue:
                        x, y = queue.popleft()
                        current += grid[x][y]
                        
                        for dx, dy in directions:
                            dx += x
                            dy += y
                            if isValid(dx, dy) and not (dx, dy) in visited:
                                visited.add((dx, dy))
                                queue.append((dx, dy))
                                
                        
                    fished = max(current, fished)

        return fished

                
                

        
    
```

```python
# https://leetcode.com/problems/walls-and-gates/?envType=problem-list-v2&envId=ahy2d3lh
from collections import deque
class Solution:
    def wallsAndGates(self, rooms: List[List[int]]) -> None:
        """
        Do not return anything, modify rooms in-place instead.
        """
        MAX = 2147483647
        rows = len(rooms)
        cols = len(rooms[0])
        queue = deque([])
        directions = [(0,1), (1,0), (0,-1), (-1,0)]
        isValid = lambda x, y: 0 <= x < rows and 0 <= y < cols

        for r in range(rows):
            for c in range(cols):
                if rooms[r][c] == 0:
                    queue.append((r,c))

        while queue:
            x, y = queue.popleft()

            for dx, dy in directions:
                dx += x
                dy += y

                if isValid(dx, dy) and rooms[dx][dy] == MAX:
                    rooms[dx][dy] = rooms[x][y] + 1
                    queue.append((dx,dy))
        
        return rooms
```

```python
# https://leetcode.com/problems/rotting-oranges/
from collections import deque
from typing import List

class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        rows = len(grid)
        cols = len(grid[0])
        adjacents = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        visited = set()
        queue = deque()

        def isValid(x: int, y: int) -> bool:
            return 0 <= x < rows and 0 <= y < cols

        fresh_orage = 0
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 2:
                    queue.append((r, c))
                if grid[r][c] == 1:
                    fresh_orage += 1
        if fresh_orage == 0:
            return 0

        minuts = -1

        while queue:
            size = len(queue)
            for _ in range(size):
                row, col = queue.popleft()
                
                for x, y in adjacents:
                    next_row, next_col = x + row, y + col
                    if isValid(next_row, next_col) and (next_row, next_col) not in visited and grid[next_row][next_col] == 1:
                        queue.append((next_row, next_col))
                        grid[next_row][next_col] = 2
                        fresh_orage -= 1
                    visited.add((next_row, next_col))

            minuts += 1
        
        return -1 if fresh_orage > 0 else minuts
"""
[2,2,2]
[0,2,2]
[1,0,2]
"""
```

```python
# Important: BFS template for word ladder problem
# https://leetcode.com/problems/word-ladder/
class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        # Create a set of all words in the word list for quick lookup.
        wordSet = set(wordList)

        # If endWord is not in the word set, no valid transformation exists.
        if endWord not in wordSet:
            return 0

        # Use a queue to perform BFS (Breadth-First Search).
        word_queue = deque()
        word_queue.append(beginWord)

        # Distance from the beginWord (initially 1 since beginWord is counted).
        distance = 1

        while word_queue:
            level_size = len(word_queue)

            for _ in range(level_size):
                current_word = word_queue.popleft()

                # If the current word is the endWord, return the distance.
                if current_word == endWord:
                    return distance

                # Try changing each character in the current word.
                for i in range(len(current_word)):
                    for c in 'abcdefghijklmnopqrstuvwxyz':
                        if c == current_word[i]:
                            continue  # Skip same character

                        new_word = current_word[:i] + c + current_word[i+1:]

                        # If the new word is in the word set, add it to the queue.
                        if new_word in wordSet:
                            word_queue.append(new_word)
                            wordSet.remove(new_word)  # Remove to prevent revisiting

            # Increment distance after processing the current level.
            distance += 1

        # If no transformation sequence leads to the endWord, return 0.
        return 0
```

```python
# https://leetcode.com/problems/all-nodes-distance-k-in-binary-tree

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque

class Solution:
    def distanceK(self, root: TreeNode, target: TreeNode, k: int) -> List[int]:
        # The key in this problem is the parent
        def dfs(node, parent):
            if node is None:
                return
            
            node.parent = parent
            dfs(node.left, node)
            dfs(node.right, node)
        
        dfs(root, None)
        queue = deque([target])
        seen = {target.val}
        distance = 0

        # Breadth First Search
        while queue and distance < k:
            current_len = len(queue)
            for _ in range(current_len):
                node = queue.popleft()
                for neightbor in [node.left, node.right, node.parent]:
                    if neightbor and neightbor.val not in seen:
                        queue.append(neightbor)
                        seen.add(neightbor.val)
            distance += 1
        
        return [node.val for node in queue]
            
```

```python
# https://leetcode.com/problems/amount-of-time-for-binary-tree-to-be-infected/description/
from collections import defaultdict, deque
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def amountOfTime(self, root: Optional[TreeNode], start: int) -> int:
        graph = defaultdict(list)

        def dfs(node):
            if not node:
                return None
            
            if node.left:
                graph[node.val].append(node.left.val)
                graph[node.left.val].append(node.val)
            if node.right:
                graph[node.val].append(node.right.val)
                graph[node.right.val].append(node.val)

            dfs(node.left)
            dfs(node.right)
        
        dfs(root)
        queue = deque([start])
        visited = set([start])
        minutes = -1

        while queue:
            size = len(queue)
            minutes += 1
            for _ in range(size):
                current = queue.popleft()

                for neigtbor in graph[current]:
                    if neigtbor not in visited:
                        queue.append(neigtbor)
                    visited.add(neigtbor)
        

        return minutes if minutes > 0 else 0
```
