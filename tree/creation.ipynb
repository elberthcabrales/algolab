```python
#Conver Sorted List to binary seach Tree
class Solution:
    def sortedListToBST(self) -> Optional[TreeNode]:
        """
            fill an array with the linked list
        """
        nums = [1,2,3,4,5,6,7]
    
        def createTree(left: int, right: int):
            if left > right:
                return None
            
            mid = (left + right) // 2
            root = TreeNode(nums[mid])
            root.left = createTree(left, mid - 1)
            root.right = createTree(mid + 1, right)

            return root
        
        return createTree(0, len(nums) - 1)

solution = Solution()
output_sortedListToBST = solution.sortedListToBST()
output_sortedListToBST.val
```

```python
# https://leetcode.com/problems/construct-binary-tree-from-inorder-and-postorder-traversal/editorial/
class Solution:
    def buildTree(self, inorder: List[int], postorder: List[int]) -> TreeNode:
        def helper(in_left: int, in_right: int) -> TreeNode:
            # if there are no elements to construct subtrees
            if in_left > in_right:
                return None

            # pick up the last element as a root
            val = postorder.pop()
            root = TreeNode(val)

            # root splits inorder list
            # into left and right subtrees
            index = idx_map[val]

            # build the right subtree
            root.right = helper(index + 1, in_right)
            # build the left subtree
            root.left = helper(in_left, index - 1)
            return root

        # build a hashmap value -> its index
        idx_map = {val: idx for idx, val in enumerate(inorder)}
        return helper(0, len(inorder) - 1)
```

```python
#height-balanced binary tree
# https://leetcode.com/problems/convert-sorted-list-to-binary-search-tree/description/
def sortedArrayToBST(nums):
    if not nums:
        return None

    # Find the middle element
    mid = len(nums) // 2

    # The middle element becomes the root
    root = TreeNode(nums[mid])

    # Recursively construct the left subtree and right subtree
    root.left = sortedArrayToBST(nums[:mid])
    root.right = sortedArrayToBST(nums[mid+1:])

    return root
nums = [-10,-3,0,5,9]
bst_root = sortedArrayToBST(nums)
bst_root.val  # Output: 0 (the root of the balanced BST)
```

```python
# https://leetcode.com/problems/construct-binary-tree-from-preorder-and-inorder-traversal
class Bst:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        self.preorder = preorder
        self.inorder = inorder
        self.index = 0

        return self.build(0, len(preorder) - 1)
    
    def build(self, lo: int, hi: int) -> Optional[TreeNode]:
        if lo > hi:
            return
        
        val = self.preorder[self.index]
        self.index += 1

        mid = self.inorder.index(val)

        root = TreeNode(val)
        root.left = self.build(lo, mid - 1)
        root.right = self.build(mid + 1, hi)

        return root
```

```python

# Create the tree for the traversal tests:
#        1
#      /   \
#     2     3
#    / \   / \
#   4   5 6   7
from typing import List, Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)


def preorder_recursive(root):
    result = []
    def helper(node):
        if not node:
            return
        result.append(node.val)      # Visit root
        helper(node.left)            # Traverse left
        helper(node.right)           # Traverse right
    helper(root)
    return result

print(preorder_recursive(root))  # Preorder: [1, 2, 4, 5, 3, 6, 7]

class BST:
    def buildTree(self, preorder: List[int]) -> Optional[TreeNode]:
        self.index = 0
        return self.build(preorder)
    
    def build(self, preorder: List[int]) -> Optional[TreeNode]:
        # Base case: no more elements
        if self.index == len(preorder):
            return None
        
        # Create node with current value
        node = TreeNode(preorder[self.index])
        self.index += 1
        
        # Build left subtree
        node.left = self.build(preorder)
        
        # Build right subtree
        node.right = self.build(preorder)
        
        return node
    
bst = BST()
bst_root = bst.buildTree([1, 2, 4, 5, 3, 6, 7])
print(preorder_recursive(bst_root))  # Preorder: [1, 2, 4, 5, 3, 6, 7]
```

```python
# https://leetcode.com/problems/serialize-and-deserialize-binary-tree/

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


"""
Aprendí:
    que el preorder es el más facil para construir un arbol solo incrementando en uno el indice
    que cuando estas serializando tambien se serializas los None
    y cuando tienes None dentro de tu arreglo en preorder
"""

class Codec:
    def __init__(self):
        self.N = "None"

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        arr = []
        def inorder(node):
            if not node:
                arr.append(self.N)
                return None

            arr.append(str(node.val))
            inorder(node.left)
            inorder(node.right)
        
        inorder(root)

        return ",".join(arr)
        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        self.index = 0
        def inorder(preorder: List[str]):
            val = preorder[self.index]
            self.index += 1
            
            if val == self.N:
                return None

            node = TreeNode(int(val))
            node.left = inorder(preorder)
            node.right = inorder(preorder)

            return node
        
        return inorder(data.split(","))

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```
