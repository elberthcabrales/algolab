## <a id="two-pointers"></a>1. Two Pointers Pattern

**When to Use:**
- Working with sorted arrays
- Finding pairs/triplets with specific sum
- Palindrome checking
- Removing duplicates
- Container with most water

**Time Complexity:** O(n)
**Space Complexity:** O(1)

```python
from typing import List

# Template 1: Same Array - Opposite Ends
def two_pointers_opposite(arr: List[int]) -> bool:
    """
    Use case: Palindrome check, Two Sum in sorted array
    """
    left, right = 0, len(arr) - 1
    
    while left < right:
        # Process elements at left and right
        if arr[left] != arr[right]:
            return False
        left += 1
        right -= 1
    
    return True

# Template 2: Same Array - Same Direction
def two_pointers_same_direction(arr: List[int]) -> int:
    """
    Use case: Remove duplicates, move zeros
    """
    slow = 0
    
    for fast in range(len(arr)):
        # Condition to keep element
        if arr[fast] != 0:  # Example: moving non-zeros
            arr[slow] = arr[fast]
            slow += 1
    
    return slow

# Template 3: Two Arrays
def merge_sorted_arrays(arr1: List[int], arr2: List[int]) -> List[int]:
    """
    Use case: Merge sorted arrays, intersection
    """
    i, j = 0, 0
    result = []
    
    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1
    
    # Add remaining elements
    result.extend(arr1[i:])
    result.extend(arr2[j:])
    
    return result

# Example: Container With Most Water
def maxArea(height: List[int]) -> int:
    left, right = 0, len(height) - 1
    max_area = 0
    
    while left < right:
        width = right - left
        max_area = max(max_area, min(height[left], height[right]) * width)
        
        # Move pointer with smaller height
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    
    return max_area
```

## <a id="sliding-window"></a>2. Sliding Window Pattern

**When to Use:**
- Subarray/substring problems
- Finding max/min in contiguous sequence
- Problems with "at most K" or "exactly K" constraints

**Time Complexity:** O(n)
**Space Complexity:** O(k) for hashmap

```python
from collections import defaultdict

# Template 1: Fixed Size Window
def fixed_window(arr: List[int], k: int) -> int:
    """
    Use case: Maximum sum of k consecutive elements
    """
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]
        max_sum = max(max_sum, window_sum)
    
    return max_sum

# Template 2: Variable Size Window
def longest_substring_k_distinct(s: str, k: int) -> int:
    """
    Use case: Longest substring with at most K distinct characters
    """
    char_count = defaultdict(int)
    left = 0
    max_length = 0
    
    for right in range(len(s)):
        # Expand window
        char_count[s[right]] += 1
        
        # Shrink window if constraint violated
        while len(char_count) > k:
            char_count[s[left]] -= 1
            if char_count[s[left]] == 0:
                del char_count[s[left]]
            left += 1
        
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Template 3: Minimum Window (harder variant)
def min_window_substring(s: str, t: str) -> str:
    """
    Use case: Minimum window substring containing all characters of t
    """
    if not s or not t:
        return ""
    
    # Character frequency in t
    target_count = defaultdict(int)
    for char in t:
        target_count[char] += 1
    
    required = len(target_count)
    formed = 0
    window_count = defaultdict(int)
    
    left = 0
    min_len = float('inf')
    min_left = 0
    
    for right in range(len(s)):
        char = s[right]
        window_count[char] += 1
        
        if char in target_count and window_count[char] == target_count[char]:
            formed += 1
        
        # Shrink window when valid
        while left <= right and formed == required:
            if right - left + 1 < min_len:
                min_len = right - left + 1
                min_left = left
            
            char = s[left]
            window_count[char] -= 1
            if char in target_count and window_count[char] < target_count[char]:
                formed -= 1
            left += 1
    
    return "" if min_len == float('inf') else s[min_left:min_left + min_len]
```

## <a id="binary-search"></a>3. Binary Search Pattern

**When to Use:**
- Sorted array search
- Find insertion position
- Find first/last occurrence
- Minimize/maximize optimization problems

**Time Complexity:** O(log n)
**Space Complexity:** O(1)

```python
# Template 1: Classic Binary Search
def binary_search(arr: List[int], target: int) -> int:
    """
    Find target in sorted array. Returns index or -1.
    """
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

# Template 2: Leftmost Position (Lower Bound)
def binary_search_leftmost(arr: List[int], target: int) -> int:
    """
    Find leftmost insertion position. Returns index in [0, len(arr)].
    Equivalent to Python's bisect_left.
    """
    left, right = 0, len(arr)
    
    while left < right:
        mid = (left + right) // 2
        
        if arr[mid] >= target:
            right = mid
        else:
            left = mid + 1
    
    return left

# Template 3: Rightmost Position (Upper Bound)
def binary_search_rightmost(arr: List[int], target: int) -> int:
    """
    Find rightmost insertion position. Returns index in [0, len(arr)].
    Equivalent to Python's bisect_right.
    """
    left, right = 0, len(arr)
    
    while left < right:
        mid = (left + right) // 2
        
        if arr[mid] > target:
            right = mid
        else:
            left = mid + 1
    
    return left

# Template 4: Binary Search on Answer Space
def split_array_largest_sum(nums: List[int], k: int) -> int:
    """
    Minimize the maximum sum when splitting array into k parts.
    Pattern: Binary search on answer space.
    """
    def is_feasible(max_sum: int) -> bool:
        parts = 1
        current_sum = 0
        
        for num in nums:
            if current_sum + num > max_sum:
                parts += 1
                current_sum = 0
            current_sum += num
            
            if parts > k:
                return False
        
        return True
    
    left = max(nums)
    right = sum(nums)
    
    while left < right:
        mid = (left + right) // 2
        
        if is_feasible(mid):
            right = mid
        else:
            left = mid + 1
    
    return left
```

## <a id="prefix-sum"></a>4. Prefix Sum + HashMap Pattern

**When to Use:**
- Count subarrays with specific sum
- Find longest subarray with sum = k
- Problems with "continuous subarray" conditions

**Key Insight:** If prefix_sum[j] - prefix_sum[i] = k, then subarray [i+1, j] has sum k

**Time Complexity:** O(n)
**Space Complexity:** O(n)

```python
from collections import defaultdict

# Template 1: Count Subarrays with Sum = K
def count_subarrays_with_sum(arr: List[int], k: int) -> int:
    """
    Count number of continuous subarrays with sum = k.
    """
    counts = defaultdict(int)
    counts[0] = 1  # Handle subarrays starting from index 0
    
    curr_sum = 0
    answer = 0
    
    for num in arr:
        curr_sum += num
        
        # Check if (curr_sum - k) exists
        answer += counts[curr_sum - k]
        
        # Record current prefix sum
        counts[curr_sum] += 1
    
    return answer

# Template 2: Longest Subarray with Sum = K
def longest_subarray_with_sum(arr: List[int], k: int) -> int:
    """
    Find longest subarray with sum = k.
    """
    first_occurrence = {0: -1}
    
    curr_sum = 0
    max_length = 0
    
    for i, num in enumerate(arr):
        curr_sum += num
        
        target = curr_sum - k
        if target in first_occurrence:
            length = i - first_occurrence[target]
            max_length = max(max_length, length)
        
        # Store first occurrence only
        if curr_sum not in first_occurrence:
            first_occurrence[curr_sum] = i
    
    return max_length

# Template 3: Subarray Sum Divisible by K
def count_subarrays_divisible_by_k(arr: List[int], k: int) -> int:
    """
    Count subarrays where sum is divisible by k.
    Key: If two prefix sums have same remainder, subarray between them is divisible by k.
    """
    counts = defaultdict(int)
    counts[0] = 1
    
    curr_sum = 0
    answer = 0
    
    for num in arr:
        curr_sum += num
        remainder = ((curr_sum % k) + k) % k  # Handle negative remainders
        
        answer += counts[remainder]
        counts[remainder] += 1
    
    return answer

# Template 4: Equal 0s and 1s (Contiguous Array)
def longest_subarray_equal_01(arr: List[int]) -> int:
    """
    Find longest subarray with equal 0s and 1s.
    Transform: treat 0 as -1, find longest subarray with sum = 0.
    """
    first_occurrence = {0: -1}
    
    curr_sum = 0
    max_length = 0
    
    for i, num in enumerate(arr):
        curr_sum += 1 if num == 1 else -1
        
        if curr_sum in first_occurrence:
            length = i - first_occurrence[curr_sum]
            max_length = max(max_length, length)
        else:
            first_occurrence[curr_sum] = i
    
    return max_length
```

## <a id="monotonic-stack"></a>5. Monotonic Stack/Queue Pattern

**When to Use:**
- Next greater/smaller element
- Sliding window maximum/minimum
- Histogram problems
- Temperature span

**Time Complexity:** O(n)
**Space Complexity:** O(n)

```python
from collections import deque

# Template 1: Next Greater Element (Monotonic Decreasing Stack)
def next_greater_elements(arr: List[int]) -> List[int]:
    """
    For each element, find the next greater element to its right.
    """
    n = len(arr)
    result = [-1] * n
    stack = []  # Store indices
    
    for i in range(n):
        # Pop smaller elements and set their next greater
        while stack and arr[stack[-1]] < arr[i]:
            idx = stack.pop()
            result[idx] = arr[i]
        
        stack.append(i)
    
    return result

# Template 2: Next Smaller Element (Monotonic Increasing Stack)
def next_smaller_elements(arr: List[int]) -> List[int]:
    """
    For each element, find the next smaller element to its right.
    """
    n = len(arr)
    result = [-1] * n
    stack = []
    
    for i in range(n):
        # Pop greater elements and set their next smaller
        while stack and arr[stack[-1]] > arr[i]:
            idx = stack.pop()
            result[idx] = arr[i]
        
        stack.append(i)
    
    return result

# Template 3: Sliding Window Maximum (Monotonic Deque)
def sliding_window_maximum(arr: List[int], k: int) -> List[int]:
    """
    Find maximum in each window of size k.
    """
    dq = deque()  # Store indices
    result = []
    
    for i in range(len(arr)):
        # Remove elements outside window
        if dq and dq[0] < i - k + 1:
            dq.popleft()
        
        # Remove smaller elements (maintain decreasing order)
        while dq and arr[dq[-1]] < arr[i]:
            dq.pop()
        
        dq.append(i)
        
        # Add to result when window is complete
        if i >= k - 1:
            result.append(arr[dq[0]])
    
    return result

# Template 4: Largest Rectangle in Histogram
def largest_rectangle_area(heights: List[int]) -> int:
    """
    Find largest rectangle in histogram.
    """
    stack = []  # Store indices
    max_area = 0
    heights.append(0)  # Add sentinel
    
    for i in range(len(heights)):
        while stack and heights[stack[-1]] > heights[i]:
            h = heights[stack.pop()]
            w = i if not stack else i - stack[-1] - 1
            max_area = max(max_area, h * w)
        
        stack.append(i)
    
    return max_area
```

## <a id="fast-slow-pointers"></a>6. Fast & Slow Pointers Pattern

**When to Use:**
- Linked list cycle detection
- Finding middle of linked list
- Palindrome linked list
- Happy number

**Time Complexity:** O(n)
**Space Complexity:** O(1)

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Template 1: Detect Cycle
def has_cycle(head: ListNode) -> bool:
    """
    Detect if linked list has a cycle.
    """
    if not head:
        return False
    
    slow = fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return True
    
    return False

# Template 2: Find Cycle Start
def detect_cycle(head: ListNode) -> ListNode:
    """
    Find the node where cycle begins.
    """
    if not head:
        return None
    
    # Detect cycle
    slow = fast = head
    has_cycle = False
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            has_cycle = True
            break
    
    if not has_cycle:
        return None
    
    # Find cycle start
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next
    
    return slow

# Template 3: Find Middle
def find_middle(head: ListNode) -> ListNode:
    """
    Find middle node of linked list.
    """
    slow = fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    return slow

# Template 4: Palindrome Linked List
def is_palindrome(head: ListNode) -> bool:
    """
    Check if linked list is a palindrome.
    """
    if not head or not head.next:
        return True
    
    # Find middle
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    # Reverse second half
    prev = None
    while slow:
        next_node = slow.next
        slow.next = prev
        prev = slow
        slow = next_node
    
    # Compare
    left, right = head, prev
    while right:
        if left.val != right.val:
            return False
        left = left.next
        right = right.next
    
    return True
```

## <a id="interval-merging"></a>7. Interval Merging Pattern

**When to Use:**
- Merge overlapping intervals
- Meeting rooms problems
- Insert intervals
- Interval scheduling

**Time Complexity:** O(n log n) for sorting
**Space Complexity:** O(n)

```python
from heapq import heappush, heappop

# Template 1: Merge Intervals
def merge_intervals(intervals: List[List[int]]) -> List[List[int]]:
    """
    Merge overlapping intervals.
    """
    if not intervals:
        return []
    
    intervals.sort()
    merged = []
    
    for start, end in intervals:
        # If overlaps with last interval, merge
        if merged and merged[-1][1] >= start:
            merged[-1][1] = max(merged[-1][1], end)
        else:
            merged.append([start, end])
    
    return merged

# Template 2: Insert Interval
def insert_interval(intervals: List[List[int]], new_interval: List[int]) -> List[List[int]]:
    """
    Insert new interval and merge if necessary.
    """
    result = []
    i = 0
    n = len(intervals)
    
    # Add all intervals before new_interval
    while i < n and intervals[i][1] < new_interval[0]:
        result.append(intervals[i])
        i += 1
    
    # Merge overlapping intervals
    while i < n and intervals[i][0] <= new_interval[1]:
        new_interval[0] = min(new_interval[0], intervals[i][0])
        new_interval[1] = max(new_interval[1], intervals[i][1])
        i += 1
    
    result.append(new_interval)
    
    # Add remaining intervals
    while i < n:
        result.append(intervals[i])
        i += 1
    
    return result

# Template 3: Meeting Rooms II (Minimum Groups/Resources)
def min_meeting_rooms(intervals: List[List[int]]) -> int:
    """
    Find minimum number of meeting rooms needed.
    Key: Maximum number of overlapping intervals at any point.
    """
    if not intervals:
        return 0
    
    intervals.sort()
    heap = []  # Store end times
    
    for start, end in intervals:
        # If earliest ending meeting finishes before current starts, reuse room
        if heap and heap[0] <= start:
            heappop(heap)
        
        heappush(heap, end)
    
    return len(heap)  # Heap size = max concurrent meetings

# Template 4: Non-overlapping Intervals (Greedy)
def erase_overlap_intervals(intervals: List[List[int]]) -> int:
    """
    Find minimum number of intervals to remove to make non-overlapping.
    Greedy: Sort by end time, keep intervals that end earliest.
    """
    if not intervals:
        return 0
    
    # Sort by end time
    intervals.sort(key=lambda x: x[1])
    
    end = float('-inf')
    count = 0
    
    for start, interval_end in intervals:
        if start >= end:
            # No overlap, keep this interval
            end = interval_end
        else:
            # Overlap, remove current interval
            count += 1
    
    return count
```

## <a id="backtracking"></a>8. Backtracking Pattern

**When to Use:**
- Generate all combinations/permutations
- Subsets problems
- N-Queens, Sudoku solver
- Word search

**Time Complexity:** O(2^n) for subsets, O(n!) for permutations
**Space Complexity:** O(n) for recursion stack

```python
# Template 1: Subsets
def subsets(nums: List[int]) -> List[List[int]]:
    """
    Generate all subsets.
    """
    result = []
    
    def backtrack(start: int, path: List[int]):
        result.append(path.copy())
        
        for i in range(start, len(nums)):
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
    
    backtrack(0, [])
    return result

# Template 2: Combinations
def combine(n: int, k: int) -> List[List[int]]:
    """
    Generate all combinations of k numbers from 1 to n.
    """
    result = []
    
    def backtrack(start: int, path: List[int]):
        if len(path) == k:
            result.append(path.copy())
            return
        
        for i in range(start, n + 1):
            path.append(i)
            backtrack(i + 1, path)
            path.pop()
    
    backtrack(1, [])
    return result

# Template 3: Permutations
def permute(nums: List[int]) -> List[List[int]]:
    """
    Generate all permutations.
    """
    result = []
    
    def backtrack(path: List[int], used: List[bool]):
        if len(path) == len(nums):
            result.append(path.copy())
            return
        
        for i in range(len(nums)):
            if used[i]:
                continue
            
            path.append(nums[i])
            used[i] = True
            backtrack(path, used)
            path.pop()
            used[i] = False
    
    backtrack([], [False] * len(nums))
    return result

# Template 4: General Backtracking
def backtracking_template():
    """
    General backtracking structure.
    """
    result = []
    
    def backtrack(state, choices):
        # Base case: found valid solution
        if is_solution(state):
            result.append(state.copy())
            return
        
        # Explore all choices
        for choice in choices:
            # Pruning: skip invalid choices
            if not is_valid(state, choice):
                continue
            
            # Make choice
            make_choice(state, choice)
            
            # Recurse
            backtrack(state, get_next_choices(state))
            
            # Undo choice (backtrack)
            undo_choice(state, choice)
    
    backtrack(initial_state, initial_choices)
    return result
```
