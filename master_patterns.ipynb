## Two Pointers <a id="two-pointers"></a>

**Use Case:** Arrays/strings where you need to compare elements from both ends or find pairs.

**Time Complexity:** O(n)  
**Space Complexity:** O(1)

```python
# Pattern 1: Two Pointers (Same Array)
def two_pointers(arr):
    left = 0
    right = len(arr) - 1
    ans = 0
    
    while left < right:
        # DO SOME LOGIC with arr[left] and arr[right]
        
        if CONDITION:
            left += 1
        else:
            right -= 1
    
    return ans

# Pattern 2: Two Pointers (Two Arrays)
def two_pointers_two_arrays(arr1, arr2):
    i = j = 0
    ans = []
    
    while i < len(arr1) and j < len(arr2):
        # DO LOGIC to decide which pointer to move
        if CONDITION:
            ans.append(arr1[i])
            i += 1
        else:
            ans.append(arr2[j])
            j += 1
    
    # Handle remaining elements
    while i < len(arr1):
        ans.append(arr1[i])
        i += 1
    
    while j < len(arr2):
        ans.append(arr2[j])
        j += 1
    
    return ans
```

## Sliding Window <a id="sliding-window"></a>

**Use Case:** Finding subarrays/substrings that satisfy certain conditions.

**Time Complexity:** O(n)  
**Space Complexity:** O(k) for hash map

```python
# Pattern 1: Variable Size Window
def sliding_window_variable(arr):
    left = ans = 0
    window_state = {}  # or set(), Counter(), etc.
    
    for right in range(len(arr)):
        # ADD arr[right] to window
        window_state[arr[right]] = window_state.get(arr[right], 0) + 1
        
        # SHRINK window while CONDITION is violated
        while WINDOW_CONDITION_BROKEN:
            # REMOVE arr[left] from window
            window_state[arr[left]] -= 1
            if window_state[arr[left]] == 0:
                del window_state[arr[left]]
            left += 1
        
        # UPDATE answer with current window
        ans = max(ans, right - left + 1)
    
    return ans

# Pattern 2: Fixed Size Window
def sliding_window_fixed(arr, k):
    window_state = {}
    ans = 0
    
    # Build initial window
    for i in range(k):
        window_state[arr[i]] = window_state.get(arr[i], 0) + 1
    
    ans = CALCULATE(window_state)
    
    # Slide the window
    for i in range(k, len(arr)):
        # ADD new element
        window_state[arr[i]] = window_state.get(arr[i], 0) + 1
        
        # REMOVE old element
        window_state[arr[i - k]] -= 1
        if window_state[arr[i - k]] == 0:
            del window_state[arr[i - k]]
        
        ans = max(ans, CALCULATE(window_state))
    
    return ans
```

## Binary Search <a id="binary-search"></a>

**Use Case:** Searching in sorted arrays or searching for a condition.

**Time Complexity:** O(log n)  
**Space Complexity:** O(1)

```python
# Pattern 1: Classic Binary Search
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

# Pattern 2: Leftmost Insert Position
def binary_search_leftmost(arr, target):
    left = 0
    right = len(arr)
    
    while left < right:
        mid = (left + right) // 2
        
        if arr[mid] >= target:  # Use '>' for rightmost
            right = mid
        else:
            left = mid + 1
    
    return left

# Pattern 3: Binary Search on Answer
def binary_search_answer(arr, k):
    def is_feasible(candidate):
        # CHECK if candidate satisfies the problem constraints
        parts = 0
        current_sum = 0
        
        for num in arr:
            if current_sum + num > candidate:
                parts += 1
                current_sum = 0
            current_sum += num
            
            if parts > k:
                return False
        
        return True
    
    left = max(arr)  # MINIMUM possible answer
    right = sum(arr)  # MAXIMUM possible answer
    
    while left < right:
        mid = (left + right) // 2
        
        if is_feasible(mid):
            right = mid  # For minimization problems
        else:
            left = mid + 1
    
    return left

def maximumCandies(self, candies: List[int], k: int) -> int:
    """
        we can't merge two piles together
        return the max pile of candies for each kid (k)
    """
    if sum(candies) < k:
        return 0
    def isFeasable(mid: int) -> bool:
        kids_served = 0
        for candy in candies:
            kids_served += (candy // mid)  # Each pile contributes `candy // mid` kids
            if kids_served >= k:  # Early exit if we already have enough kids served
                return True
        return kids_served >= k

    lo = 1
    hi = max(candies)

    while lo <= hi:
        mid = (lo + hi) // 2

        if isFeasable(mid):
            lo = mid + 1
        else:
            hi = mid - 1
    
    return hi
```

## Backtracking <a id="backtracking"></a>

**Use Case:** Generate all combinations, permutations, or explore all possible solutions.

**Time Complexity:** O(2^n) or O(n!)  
**Space Complexity:** O(n) for recursion stack

```python
# Pattern 1: Subsets/Combinations
def backtrack_subsets(nums):
    ans = []
    
    def backtrack(curr, start):
        # ADD current state to answer
        ans.append(curr[:])
        
        # EXPLORE all choices from current position
        for i in range(start, len(nums)):
            # MAKE choice
            curr.append(nums[i])
            
            # RECURSE
            backtrack(curr, i + 1)
            
            # UNDO choice (backtrack)
            curr.pop()
    
    backtrack([], 0)
    return ans

# Pattern 2: Permutations
def backtrack_permutations(nums):
    ans = []
    
    def backtrack(curr):
        # BASE CASE
        if len(curr) == len(nums):
            ans.append(curr[:])
            return
        
        # EXPLORE all unused elements
        for num in nums:
            if num not in curr:
                # MAKE choice
                curr.append(num)
                
                # RECURSE
                backtrack(curr)
                
                # UNDO choice (backtrack)
                curr.pop()
    
    backtrack([])
    return ans

# Pattern 3: General Backtracking Template
def backtrack(curr, OTHER_ARGUMENTS):
    if BASE_CASE:
        # MODIFY the answer
        return
    
    ans = 0
    for ITERATE_OVER_INPUT:
        # MODIFY the current state
        ans += backtrack(curr, OTHER_ARGUMENTS)
        # UNDO the modification of the current state
    
    return ans
```

## Dynamic Programming <a id="dynamic-programming"></a>

**Use Case:** Optimization problems with overlapping subproblems.

**Time Complexity:** O(n * state_size)  
**Space Complexity:** O(state_size)

```python
# Pattern 1: Top-Down (Memoization)
from functools import cache

def dp_top_down(arr):
    @cache
    def dp(i, STATE):
        # BASE CASE
        if i < 0 or i >= len(arr):
            return BASE_VALUE
        
        # RECURRENCE RELATION
        ans = 0
        for CHOICE in CHOICES:
            ans = max(ans, dp(i + TRANSITION, NEW_STATE) + COST)
        
        return ans
    
    return dp(START_INDEX, INITIAL_STATE)

# Pattern 2: Bottom-Up (Tabulation)
def dp_bottom_up(arr):
    n = len(arr)
    dp = [0] * (n + 1)
    
    # INITIALIZE base cases
    dp[0] = BASE_VALUE
    
    # FILL table iteratively
    for i in range(1, n + 1):
        for CHOICE in CHOICES:
            dp[i] = max(dp[i], dp[i - TRANSITION] + COST)
    
    return dp[n]

# Pattern 3: 2D DP (Grid/Two Sequences)
def dp_2d(text1, text2):
    m, n = len(text1), len(text2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # INITIALIZE base cases
    for i in range(m + 1):
        dp[i][0] = BASE_VALUE
    for j in range(n + 1):
        dp[0][j] = BASE_VALUE
    
    # FILL table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if CONDITION:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    return dp[m][n]
```

## Graph Algorithms <a id="graph-algorithms"></a>

**Use Case:** Graph traversal, pathfinding, connectivity.

**Time Complexity:** O(V + E)  
**Space Complexity:** O(V)

```python
from collections import deque, defaultdict

# Pattern 1: BFS (Shortest Path in Unweighted Graph)
def graph_bfs(graph, start):
    queue = deque([start])
    visited = {start}
    ans = 0
    
    while queue:
        size = len(queue)
        
        # Process all nodes at current level
        for _ in range(size):
            node = queue.popleft()
            
            # PROCESS node
            
            for neighbor in graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
        
        ans += 1
    
    return ans

# Pattern 2: DFS (Iterative)
def graph_dfs_iterative(graph, start):
    stack = [start]
    visited = {start}
    ans = 0
    
    while stack:
        node = stack.pop()
        
        # PROCESS node
        ans += 1
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
    
    return ans

# Pattern 3: DFS (Recursive)
def graph_dfs_recursive(graph, start):
    visited = set()
    
    def dfs(node):
        visited.add(node)
        ans = 1
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                ans += dfs(neighbor)
        
        return ans
    
    return dfs(start)

# Pattern 4: Grid DFS/BFS
def grid_dfs(grid):
    rows, cols = len(grid), len(grid[0])
    visited = set()
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    def is_valid(r, c):
        return 0 <= r < rows and 0 <= c < cols and CONDITION
    
    def dfs(r, c):
        if not is_valid(r, c) or (r, c) in visited:
            return 0
        
        visited.add((r, c))
        ans = 1
        
        for dr, dc in directions:
            ans += dfs(r + dr, c + dc)
        
        return ans
    
    result = 0
    for r in range(rows):
        for c in range(cols):
            if (r, c) not in visited and CONDITION:
                result = max(result, dfs(r, c))
    
    return result
```

## Dijkstra's Algorithm <a id="dijkstra"></a>

**Name:** Dijkstra's algorithm — single-source shortest paths for graphs with non-negative edge weights.

**Use Cases:** routing and navigation, shortest-latency paths in networks, weighted task scheduling, preprocessing shortest distances from a source in road networks or graphs where weights >= 0.

**Time Complexity:** O((V + E) log V) when implemented with a binary heap (heapq).
**Space Complexity:** O(V + E) for the adjacency list and O(V) for distance and heap structures.

**Notes / Tips:**
- Requires non-negative edge weights; if negative weights are present use Bellman-Ford or Johnson's algorithm (with reweighting).
- For deterministic / lexicographically smallest orders of equal-distance nodes, push (dist, node) and tie-break by node id or push (dist, node, extra) as needed.
- For many repeated single-source queries consider running Dijkstra from required sources or using multi-source techniques where applicable.
- Early exit: if you only need distance to a target, you can stop when the popped node is the target (its distance is final).


```python
from math import inf
from heapq import heappush, heappop

def dijkstra(n, graph, source):
    """Compute shortest distances from source to all nodes."""
    distances = [inf] * n
    distances[source] = 0
    heap = [(0, source)]

    while heap:
        curr_dist, node = heappop(heap)
        # Skip stale entries (we may have pushed better distances already)
        if curr_dist > distances[node]:
            continue

        for nei, weight in graph[node]:
            dist = curr_dist + weight
            if dist < distances[nei]:
                distances[nei] = dist
                heappush(heap, (dist, nei))

    return distances

# Example usage: adjacency list where graph[u] = [(v, w), ...]
if __name__ == '__main__':
    # small example graph (directed) with 5 nodes (0..4)
    n = 5
    graph = [
        [(1, 2), (2, 5)],  # 0 -> 1 (2), 0 -> 2 (5)
        [(2, 1), (3, 2)],  # 1 -> 2 (1), 1 -> 3 (2)
        [(3, 3)],          # 2 -> 3 (3)
        [(4, 1)],          # 3 -> 4 (1)
        []                 # 4 has no outgoing edges
    ]
    source = 0
    print('distances from', source, '->', dijkstra(n, graph, source))  # expected [0,2,3,4,5]
```

## Tree Traversals <a id="tree-traversals"></a>

**Use Case:** Binary tree problems.

**Time Complexity:** O(n)  
**Space Complexity:** O(h) where h is height

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Pattern 1: Inorder Traversal (Left -> Root -> Right)
def inorder_recursive(root):
    ans = []
    
    def dfs(node):
        if not node:
            return
        
        dfs(node.left)
        ans.append(node.val)
        dfs(node.right)
    
    dfs(root)
    return ans

# Pattern 2: Preorder Traversal (Root -> Left -> Right)
def preorder_recursive(root):
    ans = []
    
    def dfs(node):
        if not node:
            return
        
        ans.append(node.val)
        dfs(node.left)
        dfs(node.right)
    
    dfs(root)
    return ans

# Pattern 3: Postorder Traversal (Left -> Right -> Root)
def postorder_recursive(root):
    ans = []
    
    def dfs(node):
        if not node:
            return
        
        dfs(node.left)
        dfs(node.right)
        ans.append(node.val)
    
    dfs(root)
    return ans

# Pattern 4: Level Order Traversal (BFS)
def level_order(root):
    if not root:
        return []
    
    queue = deque([root])
    ans = []
    
    while queue:
        level_size = len(queue)
        current_level = []
        
        for _ in range(level_size):
            node = queue.popleft()
            current_level.append(node.val)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        ans.append(current_level)
    
    return ans

# Pattern 5: Bottom-Up DFS (Postorder style)
def bottom_up_dfs(root):
    def dfs(node):
        if not node:
            return BASE_VALUE
        
        left = dfs(node.left)
        right = dfs(node.right)
        
        # PROCESS with left and right results
        return COMBINE(left, right, node.val)
    
    return dfs(root)

# Pattern 6: Top-Down DFS (Preorder style)
def top_down_dfs(root):
    def dfs(node, state):
        if not node:
            return
        
        # PROCESS current node with state
        new_state = UPDATE(state, node.val)
        
        dfs(node.left, new_state)
        dfs(node.right, new_state)
    
    dfs(root, INITIAL_STATE)
```

## Heap/Priority Queue <a id="heap"></a>

## Linked List <a id="linked-list"></a>

**Use Case:** Linear data structure manipulation.

**Time Complexity:** O(n)  
**Space Complexity:** O(1)

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Pattern 1: Fast & Slow Pointers (Find Middle)
def find_middle(head):
    slow = fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    return slow

# Pattern 2: Reverse Linked List
def reverse_linked_list(head):
    prev = None
    curr = head
    
    while curr:
        next_temp = curr.next
        curr.next = prev
        prev = curr
        curr = next_temp
    
    return prev

# Pattern 3: Dummy Node
def dummy_node_pattern(head):
    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    curr = head
    
    while curr:
        if CONDITION:
            prev.next = curr.next
        else:
            prev = curr
        curr = curr.next
    
    return dummy.next

# Pattern 4: Merge Two Sorted Lists
def merge_two_lists(l1, l2):
    dummy = ListNode(0)
    curr = dummy
    
    while l1 and l2:
        if l1.val < l2.val:
            curr.next = l1
            l1 = l1.next
        else:
            curr.next = l2
            l2 = l2.next
        curr = curr.next
    
    curr.next = l1 if l1 else l2
    return dummy.next
```

## Trie <a id="trie"></a>

**Use Case:** Prefix matching, autocomplete, dictionary.

**Time Complexity:** O(m) where m is word length  
**Space Complexity:** O(n * m) where n is number of words

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word):
        node = self.root
        
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        
        node.is_end_of_word = True
    
    def search(self, word):
        node = self.root
        
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        
        return node.is_end_of_word
    
    def starts_with(self, prefix):
        node = self.root
        
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        
        return True
    
    # Pattern: DFS on Trie
    def dfs_on_trie(self, node, path, result):
        if node.is_end_of_word:
            result.append(path)
        
        for char, child in node.children.items():
            self.dfs_on_trie(child, path + char, result)
```

## String Patterns <a id="string-patterns"></a>

**Use Case:** Palindrome validation, anagram detection, substring search, character manipulation.

**Time Complexity:** O(n) for most operations, O(n²) for some  
**Space Complexity:** O(1) to O(n) depending on pattern

```python
# Pattern 1: Palindrome Check (Two Pointers)
def is_palindrome(s):
    left = 0
    right = len(s) - 1
    
    while left < right:
        # Skip non-alphanumeric characters
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        
        if s[left].lower() != s[right].lower():
            return False
        
        left += 1
        right -= 1
    
    return True

# Pattern 2: Anagram Check (Frequency Map)
from collections import Counter

def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    
    return Counter(s1) == Counter(s2)

# Alternative: Sort and compare
def is_anagram_sorted(s1, s2):
    return sorted(s1) == sorted(s2)

# Pattern 3: Group Anagrams
def group_anagrams(words):
    from collections import defaultdict
    groups = defaultdict(list)
    
    for word in words:
        # Use sorted string as key
        key = ''.join(sorted(word))
        groups[key].append(word)
    
    return list(groups.values())

# Pattern 4: Group Shifted Strings (Character Shift Pattern)
def group_shifted_strings(strings):
    from collections import defaultdict
    
    def get_key(word):
        # Calculate shift pattern between consecutive chars
        if len(word) == 1:
            return "0"
        
        key = []
        for i in range(1, len(word)):
            # Difference between consecutive characters (mod 26)
            diff = (ord(word[i]) - ord(word[i-1])) % 26
            key.append(str(diff))
        
        return ','.join(key)
    
    groups = defaultdict(list)
    for string in strings:
        groups[get_key(string)].append(string)
    
    return list(groups.values())
```

## Prefix Sum <a id="prefix-sum"></a>

**Use Case:** Range sum queries, subarray sum problems.

**Time Complexity:** O(n) build, O(1) query  
**Space Complexity:** O(n)

```python
# Prefix Sum Templates
# 1) Build 1D prefix array (pref[0] = 0, pref[i] = sum of arr[:i])
def build_prefix(arr: list) -> list:
    """Return prefix array of length len(arr)+1 where pref[0]=0.
    Then range sum for inclusive [l, r] (0-indexed) is pref[r+1] - pref[l].
    Time: O(n), Space: O(n)
    """
    n = len(arr)
    pref = [0] * (n + 1)
    for i, x in enumerate(arr, start=1):
        pref[i] = pref[i - 1] + x
    return pref

def range_sum(pref: list, l: int, r: int) -> int:
    """Return sum of arr[l..r] using pref built by build_prefix.
    Assumes 0 <= l <= r < n and pref has length n+1.
    """
    return pref[r + 1] - pref[l]

# Example usage:
arr = [3, 1, 4, 1, 5]
pref = build_prefix(arr)  # [0,3,4,8,9,14]
assert range_sum(pref, 1, 3) == 1 + 4 + 1 == 6

# 1b) Using itertools.accumulate for concise prefix building
from itertools import accumulate
def build_prefix_acc(arr: list) -> list:
    # returns pref with pref[0]=0 then cumulative sums
    return [0] + list(accumulate(arr))

# 2) 2D prefix sum (matrix)
def build_prefix_2d(grid: list) -> list:
    """Return 2D prefix sum array pref with shape (m+1)x(n+1).
    pref[i][j] = sum of grid[0..i-1][0..j-1].
    Query for rectangle (r1,c1) .. (r2,c2) inclusive uses inclusion-exclusion.
    Time: O(m*n) build, O(1) query.
    """
    m = len(grid)
    n = len(grid[0]) if m else 0
    pref = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            pref[i][j] = grid[i - 1][j - 1] + pref[i - 1][j] + pref[i][j - 1] - pref[i - 1][j - 1]
    return pref

def range_sum_2d(pref: list, r1: int, c1: int, r2: int, c2: int) -> int:
    # inclusive coordinates, 0-indexed on original grid
    return pref[r2 + 1][c2 + 1] - pref[r1][c2 + 1] - pref[r2 + 1][c1] + pref[r1][c1]

# 3) Difference array (range add updates, apply later)
def make_diff(arr: list) -> list:
    # diff has length len(arr)+1; after updates do prefix to materialize
    return [0] * (len(arr) + 1)

def range_add(diff: list, l: int, r: int, val: int) -> None:
    # add val to arr[l..r] (inclusive) using diff array
    diff[l] += val
    diff[r + 1] -= val

def apply_diff(diff: list) -> list:
    # convert diff back to the actual array (length = len(diff)-1)
    n = len(diff) - 1
    arr = [0] * n
    curr = 0
    for i in range(n):
        curr += diff[i]
        arr[i] = curr
    return arr

# 4) Notes / edge-cases / complexity
# - For 1D prefix sums: building O(n), query O(1). Keep pref length = n+1 for clean inclusive queries.
# - For 2D prefix sums: memory O(m*n), watch for large grids. Use int to avoid overflow in Python isn't an issue.
# - Difference array is great when you have many range updates and want to apply them in O(1) per update and O(n) to finalize.
# - If arr values or updates are large and you need modulo arithmetic (e.g., mod = 10**9+7), apply modulo on additions.

# Quick example demonstrating range updates via diff:
arr0 = [0] * 5  # base array
diff = make_diff(arr0)
range_add(diff, 1, 3, 5)  # add 5 to indices 1..3
range_add(diff, 2, 4, 2)  # add 2 to indices 2..4
arr_after = apply_diff(diff)  # -> [0,5,7,7,2]

# End of prefix-sum templates
```

## Counting <a id="counting"></a>

**Use Case:** Frequency counting, hash map patterns, element tracking, character/number frequency problems.

**Time Complexity:** O(n)  
**Space Complexity:** O(k) where k is unique elements

```python
from collections import defaultdict, Counter

# Pattern 1: Basic Frequency Counting (Dictionary/Counter)
def frequency_count(arr):
    freq = defaultdict(int)
    
    for num in arr:
        freq[num] += 1
    
    # Process frequencies
    for num, count in freq.items():
        # DO SOMETHING with num and its count
        pass
    
    return freq

# Pattern 2: Prefix Hash (Subarray Sum Equals K)
def count_subarrays_sum_k(arr, k):
    """Count subarrays with exact sum == k (works with negatives)"""
    counts = defaultdict(int)
    counts[0] = 1  # Empty prefix has sum 0
    ans = 0
    curr = 0
    
    for num in arr:
        curr += num
        # If there exists prefix p where curr - p == k
        ans += counts[curr - k]
        counts[curr] += 1
    
    return ans

# Pattern 3: Two Hash Maps (Bidirectional Mapping)
def is_isomorphic(s, t):
    """Check if two strings are isomorphic"""
    map_s = {}
    map_t = {}
    
    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]
        
        if char_s in map_s:
            if map_s[char_s] != char_t:
                return False
        if char_t in map_t:
            if map_t[char_t] != char_s:
                return False
        
        map_s[char_s] = char_t
        map_t[char_t] = char_s
    
    return True

# Pattern 4: Frequency Comparison (Close Strings)
def are_strings_close(word1, word2):
    """Check if strings are 'close' by frequency transformation"""
    if len(word1) != len(word2):
        return False
    
    freq1 = Counter(word1)
    freq2 = Counter(word2)
    
    # Same character sets
    if set(freq1.keys()) != set(freq2.keys()):
        return False
    
    # Same frequency distribution
    return sorted(freq1.values()) == sorted(freq2.values())

# Pattern 5: Track Min Distance Between Duplicates
def min_distance_between_duplicates(arr):
    """Find minimum distance to pick up matching cards"""
    last_seen = {}
    min_dist = float('inf')
    
    for i, num in enumerate(arr):
        if num in last_seen:
            min_dist = min(min_dist, i - last_seen[num] + 1)
        last_seen[num] = i
    
    return min_dist if min_dist != float('inf') else -1

# Pattern 6: Prefix Sum with Hash (Maximum Subarray Length)
def max_subarray_len_sum_k(arr, k):
    """Find maximum length subarray with sum == k"""
    prefix_sum = 0
    indices = {}
    max_len = 0
    
    for i, num in enumerate(arr):
        prefix_sum += num
        
        # Check if entire prefix sums to k
        if prefix_sum == k:
            max_len = i + 1
        
        # Check if there's a previous prefix
        if prefix_sum - k in indices:
            max_len = max(max_len, i - indices[prefix_sum - k])
        
        # Store first occurrence only (for max length)
        if prefix_sum not in indices:
            indices[prefix_sum] = i
    
    return max_len

# Pattern 7: Count with Condition (Lonely Numbers)
def find_lonely_numbers(nums):
    """Find numbers that appear once with no adjacent values"""
    freq = Counter(nums)
    lonely = []
    
    for num in freq:
        if freq[num] == 1 and (num - 1) not in freq and (num + 1) not in freq:
            lonely.append(num)
    
    return lonely

# Pattern 8: Set Operations (Find Common Element)
def smallest_common_element(matrix):
    """Find smallest common element in sorted rows"""
    common = set(matrix[0])
    
    for i in range(1, len(matrix)):
        common &= set(matrix[i])  # Intersection
    
    return min(common) if common else -1

# Pattern 9: Dynamic Frequency Update (Query Results)
def query_results_count_colors(queries):
    """Track distinct colors as balls are painted"""
    ball_color = {}  # ball -> color
    color_count = Counter()  # color -> frequency
    results = []
    
    for ball, color in queries:
        # Remove old color if ball was painted before
        if ball in ball_color:
            old_color = ball_color[ball]
            color_count[old_color] -= 1
            if color_count[old_color] == 0:
                del color_count[old_color]
        
        # Add new color
        ball_color[ball] = color
        color_count[color] += 1
        results.append(len(color_count))
    
    return results

# Pattern 10: Frequency Array (Fixed Range)
def count_with_array(s, t):
    """Use array for counting when range is known (e.g., 26 letters)"""
    freq = [0] * 26  # For lowercase a-z
    
    for char in s:
        freq[ord(char) - ord('a')] += 1
    
    for char in t:
        freq[ord(char) - ord('a')] -= 1
    
    # Check if all frequencies are 0 (anagram check)
    return all(f == 0 for f in freq)
```

## Stack & Queue <a id="stack-queue"></a>

**Use Case:** Parentheses matching, monotonic problems, sliding window maximum.

**Time Complexity:** O(n)  
**Space Complexity:** O(n)

```python
from collections import deque

# Pattern 1: Monotonic Increasing Stack
def monotonic_increasing_stack(arr):
    stack = []
    ans = [0] * len(arr)
    
    for i in range(len(arr)):
        while stack and arr[stack[-1]] >= arr[i]:
            stack.pop()
        
        # Stack top is the previous smaller element
        ans[i] = stack[-1] if stack else -1
        stack.append(i)
    
    return ans

# Pattern 2: Monotonic Decreasing Stack
def monotonic_decreasing_stack(arr):
    stack = []
    ans = [0] * len(arr)
    
    for i in range(len(arr)):
        while stack and arr[stack[-1]] < arr[i]:
            idx = stack.pop()
            ans[idx] = i - idx  # Distance to next greater
        
        stack.append(i)
    
    return ans

# Pattern 3: Valid Parentheses
def is_valid_parentheses(s):
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in mapping:
            top = stack.pop() if stack else '#'
            if mapping[char] != top:
                return False
        else:
            stack.append(char)
    
    return len(stack) == 0

# Pattern 4: Sliding Window Maximum (Monotonic Deque)
def sliding_window_maximum(arr, k):
    dq = deque()
    ans = []
    
    for i in range(len(arr)):
        # Remove elements outside window
        if dq and dq[0] == i - k:
            dq.popleft()
        
        # Maintain decreasing order
        while dq and arr[dq[-1]] < arr[i]:
            dq.pop()
        
        dq.append(i)
        
        # Add to result after first window
        if i >= k - 1:
            ans.append(arr[dq[0]])
    
    return ans
```

## Greedy <a id="greedy"></a>

**Use Case:** Optimization problems where local optimum leads to global optimum.

**Time Complexity:** O(n log n) typically  
**Space Complexity:** O(1) or O(n)

```python
# Pattern 1: Interval Scheduling (Sort by End Time)
def max_non_overlapping_intervals(intervals):
    if not intervals:
        return 0
    
    intervals.sort(key=lambda x: x[1])  # Sort by end time
    count = 1
    end = intervals[0][1]
    
    for start, interval_end in intervals[1:]:
        if start >= end:
            count += 1
            end = interval_end
    
    return count

# Pattern 2: Two Pointer Greedy
def greedy_two_pointer(arr, limit):
    arr.sort()
    left = 0
    right = len(arr) - 1
    ans = 0
    
    while left <= right:
        if arr[left] + arr[right] <= limit:
            left += 1
        right -= 1
        ans += 1
    
    return ans

# Pattern 3: Greedy with Accumulation
def greedy_accumulation(arr):
    prev = 0
    ans = 0
    
    for curr in arr:
        if curr > prev:
            ans += curr - prev
        prev = curr
    
    return ans
```

## Union Find (DSU) <a id="union-find"></a>

**Use Case:** Connected components, cycle detection, dynamic connectivity.

**Time Complexity:** O(α(n)) amortized  
**Space Complexity:** O(n)

```python
class UnionFind:
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [1] * size
        self.components = size
    
    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]
    
    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        
        if root_x == root_y:
            return False  # Already connected
        
        # Union by rank
        if self.rank[root_x] > self.rank[root_y]:
            self.parent[root_y] = root_x
        elif self.rank[root_x] < self.rank[root_y]:
            self.parent[root_x] = root_y
        else:
            self.parent[root_y] = root_x
            self.rank[root_x] += 1
        
        self.components -= 1
        return True
    
    def connected(self, x, y):
        return self.find(x) == self.find(y)
    
    def count_components(self):
        return self.components
```

## Topological Sort <a id="topological-sort"></a>

**Use Case:** Ordering tasks with dependencies — course prerequisites, build systems/package installs, job scheduling, and any situation that requires a linearization of a partial order.

**Algorithms:**
- Kahn's algorithm (BFS, in-degree queue): build in-degree array, enqueue nodes with in-degree 0, pop and reduce neighbors' in-degrees. Detects cycles if final order size != V.
- DFS-based (post-order reverse): run DFS, push nodes onto a stack after visiting children; reverse stack for topo order. Detect cycles via recursion stack (visiting state).

**Time Complexity:** O(V + E)
**Space Complexity:** O(V + E)

**Key details & tips:**
- There may be multiple valid topological orders; use a min-heap (priority queue) when a deterministic/lexicographically smallest order is required.
- A cycle implies no valid ordering — both Kahn and DFS variants can be used to detect cycles and report failure.
- When node labels are strings, map them to integer IDs to efficiently use arrays for in-degrees and visited states.
- Topological order is useful to run DP on DAGs (process nodes in topo order to relax edges once).
- For very large graphs prefer adjacency lists and iterative implementations to avoid recursion-depth issues.

**Common variants & applications:**
- Course scheduling (LeetCode 207/210), build dependency resolution, task pipelines, computing order of compilation/linking, finding longest paths in DAGs.

**Brief example (Kahn's algorithm):**
1. Compute in-degrees for all nodes and enqueue those with 0 in-degree.
2. Pop a node, append it to the order, decrement neighbors' in-degrees; enqueue neighbors whose in-degree becomes 0.
3. If the produced order has fewer than V nodes, a cycle exists and no topological order is possible.


```python
from collections import deque, defaultdict

# Pattern: Kahn's Algorithm (BFS)
def topological_sort(num_nodes, edges):
    # Build graph and compute in-degrees
    graph = defaultdict(list)
    in_degree = [0] * num_nodes
    
    for u, v in edges:
        graph[u].append(v)
        in_degree[v] += 1
    
    # Initialize queue with nodes having in-degree 0
    queue = deque([i for i in range(num_nodes) if in_degree[i] == 0])
    topo_order = []
    
    while queue:
        node = queue.popleft()
        topo_order.append(node)
        
        for neighbor in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
    
    # Check for cycle
    if len(topo_order) != num_nodes:
        return []  # Cycle detected
    
    return topo_order

# Pattern: DFS-based Topological Sort
def topological_sort_dfs(num_nodes, edges):
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
    
    visited = [0] * num_nodes  # 0: unvisited, 1: visiting, 2: visited
    stack = []
    
    def dfs(node):
        if visited[node] == 1:
            return False  # Cycle detected
        if visited[node] == 2:
            return True
        
        visited[node] = 1
        
        for neighbor in graph[node]:
            if not dfs(neighbor):
                return False
        
        visited[node] = 2
        stack.append(node)
        return True
    
    for i in range(num_nodes):
        if visited[i] == 0:
            if not dfs(i):
                return []  # Cycle detected
    
    return stack[::-1]
```
