```python
class Solution:
    def findTargetSumWaysTd(self, nums: List[int], target: int) -> int:
        n = len(nums)
        dp = [[-1 for _ in range(sum(nums) * 2  + 1)] for _ in range(n)]
        
        def fn(i: int, path: int) -> int:
            if i < 0:
                if path == target:
                    return 1
                return 0
            
            if dp[i][path] != -1:
                return dp[i][path]

            dp[i][path] = (fn(i - 1, path + nums[i]) + fn(i - 1, path - nums[i]))
            
            return dp[i][path] 

        return fn(n - 1, 0)
    
    def findTargetSumWays(self, nums: List[int], target: int) -> int:
        n = len(nums)
        total = sum(nums)
        if total < abs(target):
            return 0
            
        dp = [[0 for _ in range(2*total+1)] for _ in range(n)]
        dp[0][total + nums[0]] = 1
        dp[0][total - nums[0]] += 1

        for i in range(1, len(nums)):
            # For every possible target sum
            for t in range(-total, total+1):
                # If at least one expression (during previous iterations) evaluated to this target sum
                if dp[i - 1][total + t] > 0:
                    dp[i][total + t + nums[i]] += dp[i - 1][total + t]
                    dp[i][total + t - nums[i]] += dp[i - 1][total + t]
    
        return dp[len(nums)-1][target+total]
```

```python
# https://leetcode.com/problems/ones-and-zeroes
class Solution:
    def findMaxForm(self, strs: List[str], m: int, n: int) -> int:
        tuples = [(s.count("0"), s.count("1")) for s in strs]  # (zeros, ones)

        # dp[i][j] = max number of strings with at most i zeros and j ones
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        for zeros, ones in tuples:
            # reverse loop (like 0/1 knapsack to avoid reuse of same string)
            for i in range(m, zeros - 1, -1):
                for j in range(n, ones - 1, -1):
                    dp[i][j] = max(dp[i][j], 1 + dp[i - zeros][j - ones])

        return dp[m][n]
    
    def findMaxForm(self, strs: List[str], m: int, n: int) -> int:
        tuples = [(s.count("0"), s.count("1")) for s in strs] # tuples of zeros and ones

        memo = {}

        def fn(i: int, curr_zeros: int, curr_ones: int) -> int:
            if i < 0:
                return 0
            
            if curr_zeros < 0 or curr_ones < 0:
                return float("-inf")

            if (i, curr_zeros, curr_ones) in memo:
                return memo[(i, curr_zeros, curr_ones)]
            
            zeros, ones = tuples[i]
            taken = 0
            if zeros <= curr_zeros and ones <= curr_ones:
                taken = 1 + fn(i - 1, curr_zeros - zeros, curr_ones - ones)
            
            not_taken = fn(i - 1, curr_zeros, curr_ones)

            res = max(taken, not_taken)

            memo[(i, curr_zeros, curr_ones)] = res

            return memo[(i, curr_zeros, curr_ones)]


        return fn(len(strs) - 1, m, n)
```

```python
# https://leetcode.com/problems/maximum-profit-in-job-scheduling
class Solution:
    def jobScheduling(self, startTime: List[int], endTime: List[int], profit: List[int]) -> int:
        n = len(startTime)

        dp = [0] * (n + 1)
        jobs = list(zip(startTime, endTime, profit))
        jobs.sort(key=lambda e: e[1])

        endTime.sort()

        for i, (start, end, profit) in enumerate(jobs):

            index = bisect_right(endTime, start) - 1

            # max(taken, skiped)
            dp[i+1] = max(profit + dp[index + 1], dp[i])

        
        return dp[-1]
```

```python
class Solution:
    def minRefuelStops(self, target: int, startFuel: int, stations: List[List[int]]) -> int:
        if target <= startFuel:
            return 0

        n = len(stations)
        if n == 0 or startFuel < stations[0][0]:
            return -1

        # Determine an upper bound for fuel states (same as max possible fuel)
        max_fuel = target + sum(f for _, f in stations)

        # dp[i][f] = min stops to be at station i with exactly f fuel
        dp = [[math.inf] * (max_fuel + 1) for _ in range(n + 1)]
        dp[0][startFuel] = 0  # start before any station

        # Iterate over stations
        for i in range(n):
            pos, fuel_gain = stations[i]
            for fuel in range(max_fuel + 1):
                if dp[i][fuel] == math.inf:
                    continue

                # if we can't even reach this station, skip both choices
                if fuel < pos:
                    continue

                # skip current station
                dp[i + 1][fuel] = min(dp[i + 1][fuel], dp[i][fuel])

                # take current station
                nf = min(max_fuel, fuel + fuel_gain)
                dp[i + 1][nf] = min(dp[i + 1][nf], dp[i][fuel] + 1)

        # find answer: min stops among any fuel >= target after all stations
        ans = min(dp[n][fuel] for fuel in range(target, max_fuel + 1))
        return -1 if ans == math.inf else ans

            
            
    def minRefuelStops(self, target: int, startFuel: int, stations: List[List[int]]) -> int:
        n = len(stations)
        # dp[t] = max distance reachable with t refuel stops
        dp = [0] * (n + 1)
        dp[0] = startFuel
        
        for i, (pos, fuel) in enumerate(stations):
            # iterate backwards to avoid overwriting states from this round
            for t in range(i, -1, -1):
                if dp[t] >= pos:
                    dp[t + 1] = max(dp[t + 1], dp[t] + fuel)
        
        # Find the smallest t where we can reach target
        for t in range(n + 1):
            if dp[t] >= target:
                return t
        return -1
```

```python
class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        n = len(nums)
        total = sum(nums)

        # If sum is odd, cannot partition equally
        if total % 2 != 0:
            return False

        target = total // 2
        dp = [[False] * (target + 1) for _ in range(n + 1)]

        # Base case: sum 0 is always possible
        for i in range(n + 1):
            dp[i][0] = True

        # Build table
        for i in range(1, n + 1):  # start from 1 to avoid index error
            for t in range(1, target + 1):
                if nums[i - 1] <= t:
                    dp[i][t] = dp[i - 1][t] or dp[i - 1][t - nums[i - 1]]
                else:
                    dp[i][t] = dp[i - 1][t]

        return dp[n][target]
        

    def canPartitionBU(self, nums: List[int]) -> bool:
        n = len(nums)
        total = sum(nums)

        if total % 2 != 0:
            return False
        
        memo = {}

        def fn(i: int, target: int):

            if i < 0:
                if target == 0:
                    return True
                return False
            
            if (i, target) in memo:
                return memo[(i, target)]
            
            memo[(i, target)] = fn(i - 1, target - nums[i - 1]) or fn(i - 1, target)

            return memo[(i, target)]
        
        halve = total // 2
        return fn(n -1, halve)
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)

        if n <= 3:
            return max(nums)
        

        return max(self.helperBU(nums[1:]), self.helperBU(nums[:-1]))

    def helperBU(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [0 for _ in range(n)]
        dp[0] = nums[0]
        dp[1] = max(nums[1], nums[0])
        dp[2] = max(nums[1], nums[0] + nums[2])

        for i in range(3, n):
            taken = dp[i - 2] + nums[i]
            skip = dp[i - 1]
            dp[i] = max(taken, skip)

        
        return dp[n-1]
        
    def helper(self, nums: List[int]) -> int:
        memo = {}

        def fn(i: int) -> int:
            if i < 0:
                return 0

            if i in memo:
                return memo[i]
            
            take = nums[i] + fn(i - 2)
            skip = fn(i - 1)

            res = max(take, skip)

            memo[i] = res

            return memo[i]
        
        return fn(len(nums) - 1)
```

```python
class Solution:
    def findMaxForm(self, strs: List[str], m: int, n: int) -> int:
        tuples = [(s.count("0"), s.count("1")) for s in strs]

        # bottom-up DP table
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        for z, o in tuples:
            for i in range(m, z - 1, -1):      # go backwards
                for j in range(n, o - 1, -1):  # go backwards
                    dp[i][j] = max(dp[i][j], 1 + dp[i - z][j - o])

        return dp[m][n]

    def findMaxFormTD(self, strs: List[str], m: int, n: int) -> int:
        tuples = [(s.count("0"), s.count("1")) for s in strs] # tuples of zeros and ones

        memo = {}

        def fn(i: int, curr_zeros: int, curr_ones: int) -> int:
            if i < 0:
                return 0
            
            if curr_zeros < 0 or curr_ones < 0:
                return float("-inf")

            if (i, curr_zeros, curr_ones) in memo:
                return memo[(i, curr_zeros, curr_ones)]
            
            zeros, ones = tuples[i]
            taken = 0
            if zeros <= curr_zeros and ones <= curr_ones:
                taken = 1 + fn(i - 1, curr_zeros - zeros, curr_ones - ones)
            
            not_taken = fn(i - 1, curr_zeros, curr_ones)

            res = max(taken, not_taken)

            memo[(i, curr_zeros, curr_ones)] = res

            return memo[(i, curr_zeros, curr_ones)]


        return fn(len(strs) - 1, m, n)
```

```python
# https://leetcode.com/problems/uncrossed-lines
class Solution:
    def maxUncrossedLines(self, nums1: List[int], nums2: List[int]) -> int:
        n = len(nums1)
        m = len(nums2)
        dp = [[0 for _ in range(m + 1)] for _ in range(n + 1)]

        for r in range(1, n + 1):
            for c in range(1, m + 1):
                if nums1[r - 1] == nums2[c - 1]:
                    dp[r][c] = 1 + dp[r - 1][c - 1]
                else:
                    dp[r][c] = max(dp[r-1][c], dp[r][c-1])
        
        return dp[n][m]


    def maxUncrossedLinesTD(self, nums1: List[int], nums2: List[int]) -> int:
        n = len(nums1)
        m = len(nums2)
        memo = {}
        def fn(i: int, j: int) -> int:
            if i < 0 or j < 0:
                return 0
            if (i, j) in memo:
                return memo[(i, j)]
            
            if nums1[i] == nums2[j]:
                memo[(i, j)] = 1 + fn(i - 1, j - 1)
            else:
                memo[(i, j)] = max(fn(i - 1, j), fn(i, j - 1))

            return memo[(i, j)]
        
        return fn(n - 1, m - 1)
```

```python
# https://leetcode.com/problems/jump-game/?envType=problem-list-v2&envId=dynamic-programming

class Solution:
    def canJumptd(self, nums: List[int]) -> bool:
        n = len(nums)

        @cache
        def fn(i: int, steps: int) -> bool:
            if steps >= n - 1:
                return True
            
            for j in range(i + 1, steps + 1):
                if fn(j, j + nums[j]):
                    return True
            
            return False
        

        return fn(0, nums[0])

    def canJump(self, nums: List[int]) -> bool:
        n = len(nums)

        if n == 1:
            return True

        dp = [0] * (n + 1)
        dp[0] = nums[0]

        for i in range(1, n + 1):
            dp[i] = max(dp[i - 1] - 1, nums[i - 1])

            if dp[i] <= 0 and i < n:
                return False
        
        return True

    def canJump1(self, nums: List[int]) -> bool:
        """
            [3,2,1,0,4]

            3 - 1
                select 2 or 1
        """
        steps = 0
        n = len(nums)

        for step in range(len(nums) - 1):
            steps = max(step + nums[step], steps)

            if steps - step <= 0:
                return False

            if steps >= n - 1:
                return True
        
        return True
```

```python
from functools import cache

class Solution:
    def maxProfit(self, prices: List[int], fee: int) -> int:
        """
            hold and sale
        """
        n = len(prices)

        @cache
        def fn(day, hold: bool) -> int:
            if day == n:
                return 0
            
            # sale
            if hold:
                taken = prices[day] + fn(day + 1, False)
                skip = fn(day + 1, hold)
                
                return max(taken, skip)
            # buy
            else:
                taken = fn(day + 1, True) -  prices[day] - fee
                skip = fn(day + 1, False)

                return max(taken, skip)

        
        return fn(0, False)
    
    def maxProfit(self, prices: List[int], fee: int) -> int:
        n = len(prices)
        
        # Step 2: Create DP table
        dp = [[0] * 2 for _ in range(n + 1)]
        
        # Step 3: Base cases (already initialized to 0)
        # dp[n][0] = dp[n][1] = 0
        
        # Step 4 & 5: Fill table in reverse order
        for day in range(n - 1, -1, -1):
            # Not holding stock (hold = 0)
            # Original: taken = fn(day + 1, True) - prices[day] - fee
            #          skip = fn(day + 1, False)
            buy = dp[day + 1][1] - prices[day] - fee
            skip = dp[day + 1][0]
            dp[day][0] = max(buy, skip)
            
            # Holding stock (hold = 1)  
            # Original: taken = prices[day] + fn(day + 1, False)
            #          skip = fn(day + 1, hold)
            sell = prices[day] + dp[day + 1][0]
            skip = dp[day + 1][1]
            dp[day][1] = max(sell, skip)
        
        # Step 6: Return answer
        return dp[0][0]  # Start at day 0, not holding
```
