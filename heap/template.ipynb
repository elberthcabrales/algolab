```python
# https://leetcode.com/problems/kth-largest-element-in-a-stream/

from heapq import heappop, heappush
class KthLargest:

    def __init__(self, k: int, nums: List[int]):
        self.scores = sorted([num for num in nums])
        self.top = k
    def add(self, val: int) -> int:
        top = self.top
        scores = self.scores
        heappush(scores, val)
        while len(scores) > top:
            heappop(scores)

        return scores[0]
```

```python
# https://leetcode.com/problems/kth-largest-element-in-an-array/
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        heap = sorted(nums)

        while len(heap) > k:
            heappop(heap)

        return heap[0]
```

```python
# https://leetcode.com/problems/k-closest-points-to-origin/

from math import dist
from heapq import heappop, heappush

class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        heap = []
        
        start_point = (0, 0)
        for point in points:
            heappush(heap, (-dist(point, start_point), point))

            if len(heap) > k:
                heappop(heap)
        
        

        return [point for _, point in heap]
                
```

```python
# https://leetcode.com/problems/top-k-frequent-words/
from heapq import heappop, heappush
from collections import Counter

class Solution:
    def topKFrequent(self, words: List[str], k: int) -> List[str]:
        heap = []

        for word, freq in Counter(words).items():
            heappush(heap, (-freq, word))
        
        ans = []
        while k:
            _, word = heappop(heap)
            ans.append(word)
            k -= 1

        
        return ans
```

```python
# https://leetcode.com/problems/top-k-frequent-elements/
from collections import Counter
from heapq import heappop, heappush

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        heap = []

        for key, freq in Counter(nums).items():
            heappush(heap, (freq, key))

            if len(heap) > k:
                heappop(heap)
        

        return [key for _, key in heap]
```

```python
#https://leetcode.com/problems/maximum-number-of-points-from-grid-queries
from heapq import heappush, heappop
from typing import List

class Solution:
    def maxPoints(self, grid: List[List[int]], queries: List[int]) -> List[int]:
        rows, cols = len(grid), len(grid[0])
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        isValid = lambda x, y: 0 <= x < rows and 0 <= y < cols

        # Sort queries and maintain original indices
        sorted_queries = sorted((query, idx) for idx, query in enumerate(queries))
        hashMap = {}

        # Min-heap to process cells based on value
        heap = [(grid[0][0], 0, 0)]
        visited = set([(0, 0)])
        points = 0

        for query, _ in sorted_queries:
            # Process the grid for the current query threshold
            while heap:
                value, r, c = heap[0]
                if value >= query:
                    break
                heappop(heap)
                points += 1
                # Explore neighbors
                for dr, dc in directions:
                    nr, nc = r + dr, c + dc
                    if isValid(nr, nc) and (nr, nc) not in visited:
                        visited.add((nr, nc))
                        heappush(heap, (grid[nr][nc], nr, nc))
            hashMap[query] = points
        
        # Map the results back to the original query order
        return [hashMap[query] for query in queries]

```

```python
# https://leetcode.com/problems/process-tasks-using-servers
from heapq import heapify


class Solution:
    def assignTasks(self, servers: List[int], tasks: List[int]) -> List[int]:
        res, unavailable = [], []
        available = [(weight, id) for id, weight in enumerate(servers)]
        heapify(available)

        for time, task in enumerate(tasks):
            while unavailable and unavailable[0][0] == time:
                _, weight, id = heappop(unavailable)
                heappush(available, (weight, id))

            if len(available) > 0:
                weight, id = heappop(available)
                heappush(unavailable, (time + task, weight, id))
                res.append(id)
            else:
                finishTime, weight, id = heappop(unavailable)
                heappush(unavailable, (finishTime + task, weight, id))
                res.append(id)

        return res
```

```python
# https://leetcode.com/problems/remove-stones-to-minimize-the-total/
class Solution:
    def minStoneSum(self, piles: List[int], k: int) -> int:
        heap = sorted([-pile for pile in piles])

        while k:
            pile = -heappop(heap)

            part = pile // 2

            heappush(heap, -(pile - part))
            k -= 1

        
        return -sum(heap)
```

```python
class Solution:
    def halveArray(self, nums: List[int]) -> int:
        halve = sum(nums) / 2
        heap = [-num for num in nums]

        heapq.heapify(heap)

        accum = 0
        for i in range(len(nums)):
            num = heapq.heappop(heap) / 2
            accum += (-num)

            if accum >= halve:
                return i + 1
            
            heapq.heappush(heap, num)
```

```python
# keep min and max heaps
# https://leetcode.com/problems/find-median-from-data-stream/

from heapq import heappop, heappush
class MedianFinder:
    def __init__(self):
        self.min_heap = []
        self.max_heap = []


    def addNum(self, num: int) -> None:
        # keep sorted min_heap and max_heap
        max_heap = self.max_heap
        min_heap = self.min_heap

        heappush(max_heap, -num)
        heappush(min_heap, -heappop(max_heap))
        if len(min_heap) > len(max_heap):
            heappush(max_heap, -heappop(min_heap))
        print("min_heap: {}, max_heap: {}".format(min_heap, max_heap))

    def findMedian(self) -> float:
        max_heap = self.max_heap
        min_heap = self.min_heap

        if len(max_heap) > len(min_heap):
            return -max_heap[0]
        
        return (min_heap[0] - max_heap[0]) / 2
```

```python
"""
    https://leetcode.com/problems/stock-price-fluctuation/
    heres i an interesting pattern
    
    use a hashmap to validate that exist the same values in the heap
"""
from heapq import heappop, heappush
class StockPrice:

    def __init__(self):
        self.timestamp = 0
        self.price_map = {}
        self.min_heap = []
        self.max_heap = []

    def update(self, timestamp: int, price: int) -> None:
        self.timestamp = max(timestamp, self.timestamp)
        self.price_map[timestamp] = price
        heappush(self.min_heap, (price, timestamp))
        heappush(self.max_heap, (-price, timestamp))

    def current(self) -> int:
        return self.price_map[self.timestamp]

    def maximum(self) -> int:
        price, timestamp = self.max_heap[0]
        while -price != self.price_map[timestamp]:
            price, timestamp = heappop(self.max_heap)
        heappush(self.max_heap, (price, timestamp))

        return -price

    def minimum(self) -> int:
        price, timestamp = self.min_heap[0]
        while price != self.price_map[timestamp]:
            price, timestamp = heappop(self.min_heap)
        heappush(self.min_heap, (price, timestamp))
        return price
```

```python
"""
    https://leetcode.com/problems/find-k-pairs-with-smallest-sums/
"""
class Solution:
    def kSmallestPairs(self, nums1: List[int], nums2: List[int], k: int) -> List[List[int]]:
        from heapq import heappush, heappop
        m = len(nums1)
        n = len(nums2)

        ans = []
        visited = set()

        minHeap = [(nums1[0] + nums2[0], (0, 0))]
        visited.add((0, 0))

        while k > 0 and minHeap:
            _, (i, j) = heappop(minHeap)
            ans.append([nums1[i], nums2[j]])

            if i + 1 < m and (i + 1, j) not in visited:
                heappush(minHeap, (nums1[i + 1] + nums2[j], (i + 1, j)))
                visited.add((i + 1, j))

            if j + 1 < n and (i, j + 1) not in visited:
                heappush(minHeap, (nums1[i] + nums2[j + 1], (i, j + 1)))
                visited.add((i, j + 1))
            k = k - 1
        
        return ans
```

```python
# https://leetcode.com/problems/smallest-range-covering-elements-from-k-lists/description/
import heapq
from typing import List
import math

class Solution:
    def smallestRange(self, nums: List[List[int]]) -> List[int]:
        result = [-math.inf, math.inf]
        heap = []
        maxValue = -math.inf

        # Step 1: Initialize the heap
        for i in range(len(nums)):
            heap.append((nums[i][0], i, 0))
            maxValue = max(maxValue, nums[i][0])

        heapq.heapify(heap)

        # Step 2: Process the heap
        while True:
            low, rowIndex, colIndex = heapq.heappop(heap)

            # Update the range if smaller
            if maxValue - low < result[1] - result[0]:
                result = [low, maxValue]

            # Move to the next element in the current row
            colIndex += 1
            if colIndex == len(nums[rowIndex]):  # Row exhausted
                return result
            
            nextValue = nums[rowIndex][colIndex]
            maxValue = max(maxValue, nextValue)
            heapq.heappush(heap, (nextValue, rowIndex, colIndex))


# Test
nums = [[4,10,15,24,26], [0,9,12,20], [5,18,22,30]]
s = Solution()  # Output: [20,24]
s.smallestRange(nums)
```

```python
# https://leetcode.com/problems/two-best-non-overlapping-events/editorial/

class Solution:
    def maxTwoEvents(self, events: List[List[int]]) -> int:
        # Create a list to store the pair (end time, value) for events
        pq = []

        # Sort the events by their start time
        events.sort(key=lambda x: x[0])

        max_val = 0
        max_sum = 0

        for event in events:
            # Pop all valid events from the priority queue and take their maximum
            while pq and pq[0][0] < event[0]:
                max_val = max(max_val, pq[0][1])
                heapq.heappop(pq)

            # Calculate the maximum sum by adding the current event's value and the best previous event's value
            max_sum = max(max_sum, max_val + event[2])

            # Push the current event's end time and value into the heap
            heapq.heappush(pq, (event[1], event[2]))

        return max_sum
```

```python
from heapq import heappop, heappush
class Solution:
    def maxAverageRatio(self, classes: List[List[int]], extraStudents: int) -> float:
        pq = []

        # prioritize the improvement for all the calses
        for p, t in classes:
            improvement = ((p + 1) / (t + 1)) - (p / t)
            heappush(pq, (-improvement, p, t))
        
        # apply the imporovements and track where is the better candidate
        for _ in range(extraStudents):
            _, p, t = heappop(pq)

            p += 1
            t += 1

            improvement = ((p + 1) / (t + 1)) - (p / t)

            heappush(pq, (-improvement, p, t))


        total = 0
        
        while pq:
            _, p, t = heappop(pq)

            total += (p / t)
        
        return total / len(classes)
```

```python
from heapq import heappush, heappop
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None

        heap = []
        for i, node in enumerate(lists):
            if node:
                heappush(heap, (node.val, i, node))

        dummy = ListNode(0)
        curr = dummy

        while heap:
            val, idx, node = heappop(heap)
            curr.next = node
            curr = curr.next

            if node.next:
                heappush(heap, (node.next.val, idx, node.next))

        return dummy.next
```

```python

from heapq import heappush, heappop

class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        """
            [[2,15],[4,9],[9,29],[16,23],[36,45]]
            [2,15], [16, 23], [36, 45]
            [4,9], [9,29]
        

        """
        intervals.sort(key=lambda e: e[0])
        heap = []

        for start, end in intervals:
            # while discarting overlaped
            if heap and heap[0][0] <= start:
                room = heappop(heap)
                start = room[1]

            heappush(heap, (end, start))
        
        return len(heap)
```
